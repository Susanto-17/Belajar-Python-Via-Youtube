# Box Plot / Box Chart

Dalam sesi ini kita akan mempelajari cara membuat box plot dengan Matplotlib.

## 1. Import Modules 

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## 2. Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

## 3. Simple Box Plot

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

## 4. Notched Box Plot 

```python
plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Notched Box Plot')
plt.ylabel('Sumbu Y')
plt.xlabel('Sumbu X')

plt.grid()

plt.show()
```

## 5. Pengaturan marker dan outlier

```python
plt.boxplot(data,
           labels=['Data'],
           showfliers=True,
           flierprops={'markerfacecolor':'g', 'marker':'d'})

plt.title('Pengaturan Marker')
plt.ylabel('Sumbu Y')
plt.xlabel('Smbu X')

plt.grid()

plt.show()
```

```python
plt.boxplot(data,
           labels=['Data'],
           showfliers=True,
           flierprops={'markerfacecolor':'y', 'marker':'*'})

plt.title('Pengaturan Marker')
plt.ylabel('Sumbu Y')
plt.xlabel('Smbu X')

plt.grid()

plt.show()
```

```python
plt.boxplot(data, labels=['Data'], vert=False)

plt.title('Horzontal Box Plot')
plt.ylabel('Sumbu Y')
plt.xlabel('Sumbu X')

plt.grid()

plt.show()
```

## 6. Multiple Box Plot 

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4']
```

```python
plt.boxplot(data, labels=labels)

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
