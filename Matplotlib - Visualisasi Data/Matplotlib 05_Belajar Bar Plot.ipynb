## Bar Plot / Bar Chart 

Dalam sesi ini kita akan mempelajari cara membuat bar plot dengan Matplotlib

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## 1. Simple Bar Plot 

```python
data = [25, 45, 55, 25, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

## 2. Pengaturan Grid dan Color 

```python
data = [25, 45, 55, 25, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori,
       data,
       color='red',
       alpha=0.25)

plt.grid(linestyle='--',
        linewidth=2,
        axis='y', # Pilihan x, y, both
        alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

## 3. Horizontal Bar Plot 

```python
data = [25, 45, 55, 25, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.barh(kategori, data)

plt.grid(linestyle='--',
        linewidth=1,
        axis='x',
        alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horzontal Bar plot')

plt.show()
```

```python
# Membalikkan urutan
plt.barh(kategori[::-1], data[::-1])

plt.grid(linestyle='--',
        linewidth=1,
        axis='x',
        alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horzontal Bar plot')

plt.show()
```

## 4. Stacked Bar Plot 

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label='Data 1')
plt.bar(kategori, data2, label='Data 2', bottom=data1)

plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

## 5. Grouped Bar Plot
