# Violin Plot / Violin Chart

Dalam sesi ini kita akan mempelajari cara membuat violin plot dengan Matplotlib.

## 1. Import Modules 

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## 2. Sample Dataset 

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

## 3. Simple Violin Plot 

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.ylabel('Sumbu Y')
plt.xlabel('Sumbu X')

plt.show()
```

## 4. Pengaturan pada Violin Plot 

```python
plt.violinplot(data,
              showextrema=False,
              showmeans=False,
              showmedians=False,
              quantiles=None)

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
              showextrema=True,
              showmeans=False,
              showmedians=False,
              quantiles=None)

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
              showextrema=True,
              showmeans=True,
              showmedians=False,
              quantiles=None)

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
              showextrema=True,
              showmeans=True,
              showmedians=True,
              quantiles=None)

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
              showextrema=True,
              showmeans=True,
              showmedians=True,
              quantiles=[0.25, 0.5, 0.75])

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## 5. Horizontal Violin Plot 

```python
plt.violinplot(data, vert=False)

plt.title('Simple Violin Plot')
plt.ylabel('Sumbu Y')
plt.xlabel('Sumbu X')

plt.show()
```

## 6. Multiple Violin Plot 

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
```

```python
plt.violinplot(data)

plt.title('Multiple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
