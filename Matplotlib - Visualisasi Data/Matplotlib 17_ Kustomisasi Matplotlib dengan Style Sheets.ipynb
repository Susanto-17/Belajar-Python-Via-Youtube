# Matplotlib Customisation: Style Sheet

Dalam sesi ini kita akan mempelajari kustomisasi Matplotlib dengan Style Sheets.

## 1. Import Modules 

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## 2. Predefined Styles

Matplotlib menyediakan beberapa predefined styles.

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label = 'linier')
plt.plot(x, np.log(x), 'ro-', label = 'log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

## 4. Menerapkan Multiple Style 

```python
plt.style.use(['seaborn', 'dark_background'])
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label = 'linier')
plt.plot(x, np.log(x), 'ro-', label = 'log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

## 5. Menerapkan Temporary Styling

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

```python
# temporary
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
    
plt.show()
```
